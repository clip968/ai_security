#2.넘파이 활용

```python
import numpy as np

#리스트를 생성하고 배열로 변환하기
list1 = [1, 2, 3, 4]
a = np.array(list1)
print('a.shape: ', a.shape)
print('a[0]: ', a[0])
```

```python
#2차원 배열 생성하기
b = np.array([[1, 2, 3], [4, 5, 6]])
print('b.shape: ', b.shape)
print('b[0,0]: ', b[0,0])
print('b[0]: ', b[0])
```

```python
a = np.zeros(2)
print('a\n', a)
b = np.zeros((2,2))
print('b\n', b)
c = np.ones((2,3))
print('c\n', c)
d = np.full((2,3), 5)
print('d\n', d)
e = np.eye(3)
print('e\n', e) 
```

```python
#실수형 배열 생성하기
a = np.array([1, 2], dtype=np.float64)
print(a.dtype)
#정수형 배열로 변환하기
a_i8 = a.astype(np.int8)
print(a_i8.dtype)
```

```python
arr = np.array([[0, 1, 2], [3, 4, 5]])
print('type(arr):',type(arr))
print('arr.ndim:',arr.ndim)
print('arr.dtype:',arr.dtype)
print('arr.itemsize:',arr.itemsize)
print('arr.size:',arr.size)
print('arr.nbytes:',arr.nbytes)
print('arr.T:\n',arr.T)
print('arr.shape:',arr.shape)
```

```python
#1차원 배열 생성하기
a = np.arange(8)
print('a\n', a)
#다차원 배열로 변경하기
a.shape = (2,4)
print('shape\n', a)
#1차원 배열로 변경하기
print('flatten\n', a.flatten( ))
#resize 함수로 모양 변경하기
a.resize((4,2))
print('resize\n', a)
```

```python
a = np.array([[0, 1, 2], [3, 4, 5]])
print('a\n', a)
b = a.transpose( )
print('b\n', b)
c = a.T
print('c\n', c)
```

```python
mask = np.array([0, 1, 1, 0], dtype=bool)
print(mask)
data = np.random.randn(4,2)
print('\ndata 출력\n',data) #랜덤 데이터 배열 data는 책과 다르게 나타남
print('\n마스킹된 데이터 출력\n',data[mask]) # 1, 2번 행 출력
print('\n마스킹 역전된 데이터 출력\n',data[~mask]) 
```

```python
posit = data[data > 0]
print('양수 데이터 출력', '\n',posit)
#다중 조건
over1 = data[1][data[1] > 0]
print('두 번째 행의 양수 데이터 출력','\n',over1)
```

```python
a = np.arange(6)
b = a # 얕은 복사
print(a)
print(b is a)
b[0] = 10
print(a)
```

```python
a = np.arange(6)
c = a.copy() #깊은 복사
print(c is a) # 깊은 복사하면 서로 다르다고 판단함
c[0] = 20
print('A: ', a)
print('C: ', c)
```

```python
a = np.array([3, 2, 5, 1, 4])
print('원본\n',a)
print('정렬 후\n',np.sort(a)) # 정렬 해도 원본은 아직 변하지 않음
print('원본\n',a)
print('정렬한 인덱스\n',np.argsort(a))
a.sort( )
print('원본\n',a)
```

```python
list1 = list([['허준호', '남자', 30, 183],
['이가원', '여자', 24, 162],
['배규민', '남자', 23, 179],
['고고림', '남자', 21, 182],
['이새봄', '여자', 28, 160],
['이보람', '여자', 26, 163],
['이루리', '여자', 24, 157],
['오다현', '여자', 24, 172]])
col_names = ['이름', '성별', '나이', '키']
df = pd.DataFrame(list1, columns=col_names)
```

```python
#데이터프레임의 값 연산
#성별 평균 키
mean_by_gender = df.groupby(by=['성별'], as_index=False)['키'].mean() # 성별로 그룹을 묶어서 키의 평균값을 구하는 과정
mean_by_gender.rename(columns={'키':'평균 키'}, inplace=True) 
#성별 키의 표준편차
std_by_gender = df.groupby(by=['성별'], as_index=False)['키'].std( ) # 성별로 그룹을 묶어서 키의 표준편차를 구하는 과정
std_by_gender.rename(columns = {'키':'키의 표준편차'}, inplace=True)
new_df = pd.merge(mean_by_gender, std_by_gender)
new_df
```
